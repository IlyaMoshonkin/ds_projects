## 4.2 Увеличение выборки <a name="upsampling"></a>

Напишем функцию для увеличения выборки. Проверим как изменятся метрики. 

```python
#посморим баланс классов в обучающей выборке 
target_train.value_counts(normalize=True)
```

Для увеличения выборки умножим на 4 объекты класса с меткой `1`. 

```python
from sklearn.utils import shuffle

#создадим функцию для увеличения меньшего класса в обучающей выборке
def upsample(features_train, target_train, repeat):
    features_zeros = features_train[target_train == 0]
    features_ones = features_train[target_train == 1]
    target_zeros = target_train[target_train == 0]
    target_ones = target_train[target_train == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=123)
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

#проверим баланс классов
target_upsampled.value_counts(normalize=True)
```

### Модель логистической регрессии после upsampling

```python
#обучим модель
lr = LogisticRegression(solver='liblinear', random_state=123)
lr.fit(features_upsampled, target_upsampled)

#сделаем предсказание
predictions = lr.predict(features_valid)
proba_one_valid = lr.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

После upsampling модель показывает метрики лучше чем первая модель на несбалансированной выборке. 
Значение метрик сравнимы с моделью со взвещенными классами. 

### Модель дерева решений после upsampling

```python
#обучим модель
tree = DecisionTreeClassifier(random_state=123)
tree.fit(features_upsampled, target_upsampled)

#сделаем предсказание
predictions = tree.predict(features_valid)
proba_one_valid = tree.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

Модель дерева решений после upsampling работает хуже, чем две предыдущие модели решаюшего дерева. 

### Модель случайного леса после upsampling

```python
#обучим модель
rfc = RandomForestClassifier(n_estimators = 120, random_state=123)
rfc.fit(features_upsampled, target_upsampled)

#сделаем предсказание
predictions = rfc.predict(features_valid)
proba_one_valid = rfc.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

Модель работает лучше. Немного снизилась точность, но остальные метрики выше. 

## 4.3 Уменьшение выборки <a name="downsampling"></a>

Проверим как измениятся метрики после уменьшения выборки

```python
#создадим функцию для уменьшения большего класса в обучающей выборке
def downsample(features_train, target_train, fraction):
    features_zeros = features_train[target_train == 0]
    features_ones = features_train[target_train == 1]
    target_zeros = target_train[target_train == 0]
    target_ones = target_train[target_train == 1]
    
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=123)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=123)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=123)
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)

#проверим баланс классов
target_downsampled.value_counts(normalize=True)
```

### Модель логистической регрессии после downsampling

```python
#обучим модель
lr = LogisticRegression(solver='liblinear', random_state=123)
lr.fit(features_downsampled, target_downsampled)

#сделаем предсказание
predictions = lr.predict(features_valid)
proba_one_valid = lr.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

Метрики почти одинаковы для моделей логистической регрессии после upsampling и downsampling.

### Модель дерева решений после downsampling

```python
#обучим модель
tree = DecisionTreeClassifier(random_state=123)
tree.fit(features_downsampled, target_downsampled)

#сделаем предсказание
predictions = tree.predict(features_valid)
proba_one_valid = tree.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

После downsampling F1-мера осталась на преджднем уровне в сравнении с upsampling. Увеличелось значение полноты, но уменьшилось значение точности. 

### Модель случайного леса после downsampling

```python
#обучим модель
rfc = RandomForestClassifier(n_estimators = 120, random_state=123)
rfc.fit(features_downsampled, target_downsampled)

#сделаем предсказание
predictions = rfc.predict(features_valid)
proba_one_valid = rfc.predict_proba(features_valid)[:,1]

#посмотрим значение метрик
display_metrics(target_valid, predictions)
```

# 5. Оптимизация гипперпараметров <a name="optimize"></a>

Оптимизацию проведем в два этапа:

1. Сделаем поиск лучших гипперпараметров на трех разных моделях с помощью `RandomizedSearchCV`. 
2. Из трех моделей выберем лучшую по F1-мере и оптимизируем выбранную модель с помощью `GridSearchCV`.

```python
#увеличим валидационную выборку, чтобы дальше объединить с тренировочной и обучасться на них
features_valid_upsampled, target_valid_upsampled = upsample(features_valid, target_valid, 4)
target_valid_upsampled.value_counts(normalize=True)
```

```python
%%time
rfc = RandomForestClassifier(random_state=123)

#определим диапазон гиперпараметров, из которых будем искать наилучшие
rfc_params = {'n_estimators': [120, 300, 500, 800],
                 'max_depth': range(2,30,5),
                 'min_samples_split': [2, 5, 10, 15, 100],
                 'min_samples_leaf': range(2,10,5),
                 'max_features': ['sqrt', 'log2', None],
                 'bootstrap': [True, False]}

#сделаем случайный поиск наилучших гиперпараметров
rfc_rs = RandomizedSearchCV(rfc, rfc_params, scoring='f1', cv = 3, verbose = 1, n_jobs=-1, random_state=123)
#объединим тренировочную и валидационную выборки, модель посчитает метрику на кросс-валидации
rfc_rs.fit(pd.concat([features_upsampled, features_valid_upsampled]), pd.concat([target_upsampled, target_valid_upsampled]))

print('Лучшие гипперпараметры:', rfc_rs.best_params_)
print('F1:', rfc_rs.best_score_)
```

Посмотрим таблицу с лучшими параметрами. Ранжированные наверху параметры будем использовать для перебора по сетке. 

```python
#посмотрим таблицу полученных результатов на модели случайного леса
pd.DataFrame(rfc_rs.cv_results_).drop(['mean_fit_time', 'std_fit_time', 'mean_score_time', 'std_score_time',
                                      'split0_test_score', 'split1_test_score', 'split2_test_score',
                                       'std_test_score', 
                                       'params'], axis=1).sort_values('rank_test_score').reset_index(drop=True)
```

```python
%%time

forest = RandomForestClassifier(random_state=123)

forest_params = {'n_estimators': [300, 500],
                 'max_depth': [17, 27],
                 'min_samples_split': [2, 5],
                 'min_samples_leaf': [2, 7],
                 'max_features': ['sqrt', 'log2'],
                 'bootstrap': [False]}

#сделаем поиск гипперпараметров по сетке
forest_gs = GridSearchCV(forest, forest_params, scoring='f1', cv = 3, verbose = True, n_jobs=-1)
#объединим тренировочную и валидационную выборки, модель посчитает метрику на кросс-валидации
forest_gs.fit(pd.concat([features_upsampled, features_valid_upsampled]), pd.concat([target_upsampled, target_valid_upsampled]))
print('Лучшие гипперпараметры:', forest_gs.best_params_)
print('F1:', forest_gs.best_score_)
```

Полученную модель проверим на тестовой выборке. 

```python
forest_gs.best_params_
```

# 6. Тестирование модели <a name="testing"></a>

```python
#определим лучшую модель
best_model = forest_gs.best_estimator_

#сделаем предсказание
predictions = best_model.predict(features_test)
proba_one_valid = best_model.predict_proba(features_test)[:,1]

#посмотрим значение метрик
display_metrics(target_test, predictions)
```
