# 2. Обучение модели

```python
def load_train(path):
  """
  функция для загрузки трейн датасета
  path: путь к датасету
  возвращает трейн датасет

  """
  train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.25,
    horizontal_flip=True,
    rotation_range=15)

  train_datagen_flow = train_datagen.flow_from_dataframe(
    dataframe=labels_train,
    directory=path,
    x_col='file_name',
    y_col='real_age',
    target_size=(224,224),
    class_mode='raw',
    batch_size=32,
    subset='training',
    seed=12345)

  return train_datagen_flow

def load_test(path, dataframe):
  """
  функция для загрузки тест датасета
  path: путь к датасету
  возвращает тест датасет

  """
  val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.25)

  val_datagen_flow = val_datagen.flow_from_dataframe(
    dataframe=dataframe,
    directory=path,
    x_col='file_name',
    y_col='real_age',
    target_size=(224,224),
    class_mode='raw',
    batch_size=32,
    subset='validation',
    seed=12345)
  
  return val_datagen_flow

def create_model(input_shape):
  """
  функция для создания модели
  input_shape: входная размерность выборки
  возвращает модель с задаными внутри функции параметрами
  """
  
  backbone = ResNet50(input_shape=input_shape, weights='imagenet', include_top=False)
  
  model = Sequential()
  model.add(backbone)
  model.add(GlobalAvgPool2D())
  model.add(Dense(1, activation='relu'))

  optimizer = Adam(lr=0.0001)
  model.compile(
      optimizer=optimizer,
      loss='mse',
      metrics=['mae'])
  
  return model


def train_model(model, train_data, test_data, batch_size, epochs, steps_per_epoch, validation_steps):
  """
  функция для обучения модели
  model: модель
  train_data: трейн датасет
  test_data: тест датасет
  batch_size: размер батча
  epochs: количество эпох 
  steps_per_epoch: количество обучающих объектов за эпоху
  validation_steps: количество валидационных объектов за эпоху
  возвращает обученную модель
  """
  np.random.seed(12345)
  model.fit(train_data, 
            validation_data=test_data,
            batch_size=batch_size,
            epochs=epochs,
            steps_per_epoch=steps_per_epoch,
            validation_steps=validation_steps,
            verbose=2)
  
  return model
```

```python
#загружаем выборки
train_datagen_flow = load_train('/content/drive/MyDrive/datasets/appa-real-release/train/')
val_datagen_flow = load_test('/content/drive/MyDrive/datasets/appa-real-release/valid/', labels_val)
test_datagen_flow = load_test('/content/drive/MyDrive/datasets/appa-real-release/test/', labels_test)
```

```python
features, target = next(train_datagen_flow)

#выводим 12 изображений
fig = plt.figure(figsize=(10,10))
for i in range(12):
    fig.add_subplot(3, 4, i+1)
    plt.imshow(features[i])
    plt.title(f'Age: {target[i]} years')
    plt.xticks([])
    plt.yticks([])
    plt.tight_layout()
```

```python
#создаем модель
with tf.device('/device:GPU:0'):
  model = create_model((224,224,3))
model.summary()
```

```python
#размеры тренировачного, валидационного датасетов
len(train_datagen_flow), len(val_datagen_flow)
```

```python
#обучаем модель
with tf.device('/device:GPU:0'):
  history = train_model(
    model=model,
    train_data=train_datagen_flow,
    test_data=val_datagen_flow,
    batch_size=16,
    epochs=15,
    steps_per_epoch=97,
    validation_steps=12
    )
```

```python
with tf.device('/device:GPU:0'):
  model.evaluate(test_datagen_flow, verbose=2)
```

# 3. Анализ модели

Полученна *MAE* на тестовой выборке: `7.46`. Сравним построенную модель с константной моделью.

```python
#проверим ошибку константной модели
model_const = pd.Series(labels_test['real_age'].mean(), index=labels_test.index)
mae_const = mean_absolute_error(labels_test['real_age'], model_const)
print('MAE константной модели:', mae_const.round(3))
```

*MAE* константной модели выше. Построенная модель прошла проверку на адекватность.
